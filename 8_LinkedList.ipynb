# Linked List

**What is a Linked List?**

Linked List is a form of a sequential collection and it does not have to be in order. A Linked list is made up of independent nodes that may contain any type of data and each node has a reference to the next node in the link.

### Linked List in Memory

The Linked List is a linear data structure, in which the elements are not stored at contiguous memory locations. The elements in a linked list are linked using pointers. It is implemented on the heap memory rather than the stack memory.

- Creating **Head** and **Tail**, initialize with null.

- Create a blank Node and assign a value to it and reference to null.

- Link Head and Tail with these Node.

````python
class Node:
  def __init__(self, value=None):
    self.value =value
    self.next = None

class SLinkedList:
  def __init__(self):
    self.head = None   # -----------------> O(1)
    self.tail = None   # -----------------> O(1)

  def __iter__(self):
    node = self.head
    while node:
      yield node
      node = node.next



singlyLinkList = SLinkedList()

node1 = Node(1)
node2 = Node(2)
node3 = Node(3)
node4 = Node(4)

singlyLinkList.head = node1
singlyLinkList.head.next = node2    # ---------> O(1)
singlyLinkList.head.next.next = node3
singlyLinkList.head.next.next.next = node4
singlyLinkList.tail = node4

# Time Complexity: O(1)
# Space Complexity: O(1)
````

````python
print([node.value for node in singlyLinkList])
````

````python
class Node:
  def __init__(self, value=None):
    self.value =value
    self.next = None

class SLinkedList:
  def __init__(self):
    self.head = None   # -----------------> O(1)
    self.tail = None   # -----------------> O(1)

  def __iter__(self):
    node = self.head
    while node:
      yield node
      node = node.next

  # Inserting New Element
  def insertSLL(self, value, location=-1):
    newNode = Node(value)
    if self.head is None:
      self.head = newNode  # -----------> O(1)
      self.tail = newNode
    else:
      if location == 0:
        newNode.next = self.head
        self.head = newNode
      elif location == -1:
        newNode.next = None
        self.tail.next = newNode    # ---------> O(1)
        self.tail = newNode
      else:
        temNode = self.head
        index = 0
        while index < location -1:
          tempNode = temNode.next   # --------> O(n)
          index += 1
        nextNode = tempNode.next
        tempNode.next = newNode
        newNode.next = nextNode

        if tempNode == self.tail:
          self.tail = newNode

  # Traversing trough SinglyLinked List

  def traverseSL(self):
    if self.head is None:
      print("Singly Linked List are not exits.")
    else:
      node = self.head
      while node is not None:
        print(node.value)
        node = node.next

  # Search In SinglyLinked List
  def searchSLL(self, value):
    if self.head is None:
      print("Singly Linked List are not exits.")
    else:
      node = self.head
      while node is not None:
        if node.value == value:
          print(f"here is you searched {node.value}")
          break
        node = node.next
      else:
        print(f"{value} is not found!")

  # Deletion In SinglyLinked List
  def deleteSLL(self, location):
    if self.head is None:
      print("Singly linked List is not present.")
    else:
      if self.head.next == None:
        if location == 0 or location == -1:
          self.head = self.head.next
        else:
          print("It has only one value")
      else:
        if location == 0:
          self.head = self.head.next
        elif location == -1:
          tempNode = self.head
          index = 0
          while tempNode is not None:
            if tempNode.next.next == None:
              self.tail = tempNode
              break
            tempNode = tempNode.next
          self.tail = tempNode
          self.tail.next = None

        else:
          tempNode = self.head
          index = 0
          while (index < location-1):
            tempNode = tempNode.next
            index += 1
          self.tail = tempNode
          self.tail.next = self.tail.next.next

  # Delete entire SLL
  def deleteEntire(self):
    if self.head is None:
      print("The SLL does not exist.")
    else:
      self.head = None
      self.tail = None
    # Time complexity : O(1)
    # Space Complexity : O(1)


singlyLinkList = SLinkedList()
singlyLinkList.insertSLL(0,0)
singlyLinkList.insertSLL(1,0)
singlyLinkList.insertSLL(2,0)
singlyLinkList.insertSLL(3,0)
singlyLinkList.insertSLL(10,-1)
singlyLinkList.insertSLL(11, 2)

print("Before")
print([node.value for node in singlyLinkList])


# singlyLinkList.searchSLL(11)
# singlyLinkList.deleteSLL(4)
singlyLinkList.deleteEntire()

print("After")
print([node.value for node in singlyLinkList])

````
